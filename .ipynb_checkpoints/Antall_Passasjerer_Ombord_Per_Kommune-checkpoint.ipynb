```python
import pandas as panda 
import numpy as np
from datetime import datetime
from matplotlib import pyplot as plt
```

```python
#Oppretter hoved dataframe
main_df = panda.read_csv('Ruter_data.csv',delimiter=';')
```

```python
#Lager ny tabell for kommuner og passasjerer ombord i de kommunene
passasjerer_kommuner = main_df[['Kommune','Passasjerer_Ombord']]
```

```python
passasjerer_kommuner
```

```python
#Lager ny tabell gruppert etter kommuner og summerer antall passasjerer ombord per kommune og sorterer stigende
groupkommune = passasjerer_kommuner.groupby('Kommune')['Passasjerer_Ombord'].sum().sort_values(ascending=True)
```

```python
groupkommune
```

```python
#Her plottes grafen
plt.xlabel('Antall passasjerer ombord')
groupkommune.plot(figsize=(20,20),kind='barh')
```
