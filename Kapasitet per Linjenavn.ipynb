```python
import pandas as panda 
from matplotlib import pyplot as plt
```

```python
dataframe = panda.read_csv('Ruter_data.csv',delimiter=';')
dataframe.head()
```

```python
dataframe.head(50)
dataframe.xticks(1)
```

```python
dataframe.plot.scatter(x='Kjøretøy_Kapasitet',y='Linjenavn',title='Kapasitet per Linjenavn', grid=True,figsize=(20,20))
```

```python
threshold_max = dataframe['Kjøretøy_Kapasitet'].quantile(0.90)
threshold_max
```

```python
dataframe[dataframe['Kjøretøy_Kapasitet']>threshold_max]
```

```python
threshold_min = dataframe['Kjøretøy_Kapasitet'].quantile(0.10)
threshold_min
```

```python
dataframe[dataframe['Kjøretøy_Kapasitet']<threshold_min]
```

```python
new_dataframe=dataframe[(dataframe['Kjøretøy_Kapasitet']<threshold_max) & (dataframe['Kjøretøy_Kapasitet']>threshold_min)] 
```

```python
new_dataframe.head(50)
```

```python
new_dataframe.set_xticks(1)
new_dataframe.plot.scatter(x='Kjøretøy_Kapasitet',y='Linjenavn',title='Kapasitet per linjenavn', grid=True,figsize=(15,15))
```
